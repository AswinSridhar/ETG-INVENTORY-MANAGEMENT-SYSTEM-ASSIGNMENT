```python
###Adding a product to the Inventory
import time
import json
print("#########################################")
print("***********1.Add product*****************")
print("***********2.Remove product**************")
print("***********3.Show items******************")
print("***********4.Shopping********************")
print("Note: Plz visit show items before going for\nshopping to make sure that product is avilable")
user_choice=int(input("Enter your choice: "))
if(user_choice == 1):
  id=input(print("$Enter the  id: " ))
  name=input(print("$Enter the ProductName: "))
  quat=int(input(print("$Enter the Quantity: ")))
  price=int(input(print("$Enter the Price: "))) 
  mfd=input(print("$Enter the ManufacturingDate: ")) 
  exp=input(print("$Enter the ExpiryDate: "))
  newproduct={id:{'ProductName':name , 'Quantity':quat , 'Price':price ,'ManufacturingDate':mfd , 'ExpiryDate':exp}}
  fd=open("inventory.json",'r')
  txt = fd.read()
  fd.close()
  inventory=json.loads(txt)
  inventory.update(newproduct)
  js=json.dumps(inventory)
  fd=open("inventory.json",'w')
  fd.write(js)
  fd.close()
  print("The product updated successfully")
elif (user_choice == 2):
  delid=input(print("$Enter  id to be deleted"))
  fd=open("inventory.json",'r')
  txt = fd.read()
  fd.close()
  inventory=json.loads(txt)
  del inventory[delid]
  js=json.dumps(inventory)
  fd=open("inventory.json",'w')
  fd.write(js)
  fd.close()
  print("product removed successfully")
elif (user_choice == 3):
  fd=open("inventory.json",'r')
  txt = fd.read()
  fd.close()
  inventory=json.loads(txt)
  print("&&&&&& List of products &&&&&&&")
  for i in inventory.keys():
       print("&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&")
       print(" id: ",i)
       print("ProductName: ",inventory[i]['ProductName'])
       print("Quantity available: ",inventory[i]['Quantity'])
       print("Price: ",inventory[i]['Price'])
elif (user_choice == 4):
  coname=input("Enter your name: ")
  conum=input("Enter your Ph.no: ")
  proid=input("Enter  id: ")
  qua=int(input("Enter Product Quantity: "))
  fd=open("customer.json",'r')
  text = fd.read()
  fd.close()
  customers=json.loads(text)
  fd=open("inventory.json",'r')
  txt = fd.read()
  fd.close()
  inventory=json.loads(txt)
  points=(inventory[proid]['Price']*qua)/10
  for i in customers.keys():
      if (conum == i ):
        customers[i]['points']= int(customers[i]['points']+points)
        b=0
      else:
        newcustomer={conum:{'name':coname,'points':int(points)}}
        b=1
  if(b==1):
    {
     customers.update(newcustomer)
    }

  hh=json.dumps(customers)
  ha=open("customer.json",'w')
  ha.write(hh)
  ha.close()
  fd=open("customer.json",'r')
  text = fd.read()
  fd.close()
  customerss=json.loads(text)
  for i in inventory.keys():
      if(proid == i):
          inventory[i]['Quantity']=inventory[i]['Quantity']-qua 
          print("\n\n\n")
          print("############################################")
          print("############## MORE MARKET ##################")
          print("############# Welcomes you #################")
          print("Your shopping details: ")
          print(" ID:         ",proid)
          print("Product Name:       ",inventory[i]['ProductName'])
          print("Product Price:      ",inventory[i]['Price'])
          print("Product Quantity:   ",qua)
          print("Manufacturing date: ",inventory[i]['ManufacturingDate'])
          print("Expiry date:        ",inventory[i]['ExpiryDate'])
          print("------------------------------------------")
          print("Billing amount:     ",inventory[proid]['Price']*qua)
          print("------------------------------------------")
          print("Your wallet point:  ",customerss[conum]['points'])
          print("Billing time:       ",time.ctime())
          print("#############################################")
          print("********HAPPY SHOPPING ",coname,"********")
          fd=open("sale.json",'r')
          mxt = fd.read()
          fd.close()
          sale=json.loads(mxt)
          newsale={str(len(sale)+1):{'id':proid,'ProductName':inventory[i]['ProductName'],'Quantity':qua,'Amount':inventory[proid]['Price']*qua}}
          sale.update(newsale)
          js=json.dumps(sale)
          fd=open("sale.json",'w')
          fd.write(js)
          fd.close()
          ms=json.dumps(inventory)
          ww=open("inventory.json",'w')
          ww.write(ms)
          ww.close()


```
